# Background

Souparno has cleaned up the alignment FASTA files. Problem here was that there are gaps at the ends, and they confound our analyses. Therefore, he has removed sequences that do not have these gaps at the ends, and then trimmed out the gaps from the rest of the sequences.

Here, I will use these sequences to plot sequence diversity.

```python
from Bio import AlignIO
from sdi import sdi

import pandas as pd
import seaborn
from collections import Counter
```

```python
# Open alignment file
alignment = AlignIO.read('Segment 5 NP Aligned and Trimmed First and Last 200 nt.fasta', 'fasta')
alignment
```

```python
# Grab the first 200 positions in the alignment - variable name may be a bit confusing, but that's cool.
first_100_matrix = numpy.array([list(rec.seq.reverse_complement()) for rec in alignment[:,-200:]], np.character)
first_100_matrix
```

```python
# Grab the last 200 positions in the alignment - variable name may be a bit confusing, but that's cool.
last_100_matrix = numpy.array([list(rec.seq.reverse_complement()) for rec in alignment[:,0:200]], np.character)
last_100_matrix
```

```python
# Custom function to get a list of entropy.
"""
Input:
- multiple sequence alignment cast as a numpy array matrix from above.

Output:
- a list of tuples, each tuple being (position, entropy)
"""
def get_entropy_list(matrix):
    entropy = []
    for n, row in enumerate(matrix.T):
        count = Counter()
        for letter in row:
            count[str(letter[0])] += 1
        entropy.append((n, sdi(count)))
    return entropy
```

```python
# Get the entropy list for first 200 positions
first_100_entropy = get_entropy_list(first_100_matrix)
```

```python
# Cast the entropy list as a dataframe for ease of plotting.
df_first_100 = pd.DataFrame(first_100_entropy)
df_first_100.columns = ['Position', 'Entropy']
df_first_100.plot(0, figsize=(10,2))

fig = plt.gcf()
fig.savefig('First 100 Entropy.pdf')
```

```python
# Everything below follows the logic for the first 200.
last_100_entropy = get_entropy_list(last_100_matrix)
```

```python
df_last_100 = pd.DataFrame(last_100_entropy)
df_last_100.columns = ['Position', 'Entropy']
df_last_100.plot(0, figsize=(10,2))

fig = plt.gcf()
fig.savefig('Last 100 Entropy.pdf')
```

```python
df_all = df_first_100.append(df_last_100)
df_all.reset_index(inplace=True)
del df_all['index']
del df_all['Position']
df_all.plot(figsize=(14,2))

fig = plt.gcf()
fig.savefig('First and Last 200 Entropy.pdf')
```
