```python
from Bio import AlignIO
from Bio.Seq import reverse_complement
import pandas as pd
import seaborn
import numpy as np
```

```python
segments = ['0', '1 PB2', '2 PB1', '3 PA', '4', '5 NP', '6', '7 M1', '8 NS1']
number = 5
segment = segments[number]
```

```python
print segment
```

```python
alignment = AlignIO.read('Segment %s Aligned and Trimmed.fasta' % segment, 'fasta')
alignment
```

```python
alignment_matrix = numpy.array([list(rec.seq.reverse_complement()) for rec in alignment], np.character)
alignment_matrix
```

```python
from collections import Counter
from sdi import sdi

def get_entropy_list(matrix):
    entropy = []
    for n, row in enumerate(matrix.T):
        count = Counter()
        for letter in row:
            count[str(letter[0])] += 1
        entropy.append((n, sdi(count)))
    return entropy
```

```python
entropy = get_entropy_list(alignment_matrix)
```
